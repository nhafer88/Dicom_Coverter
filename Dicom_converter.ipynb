```python
import numpy as np
```

```python
import pydicom
```

```python
from PIL import Image
```

```python
ds = pydicom.dcmread('DICOM IMAGE/US_6375725dbfef743b.dcm')
```

```python
new_image = ds.pixel_array.astype(float)
```

```python
scaled_image = (np.maximum(new_image,0)/new_image.max())*255.0 #float pixels
```

```python
scaled_image = np.uint8(scaled_image) #integer pixels
```

```python
final_image = Image.fromarray(scaled_image)
```

```python
final_image
```

```python
final_image.save('test.jpg')
```
